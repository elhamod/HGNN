```python
%matplotlib inline

import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
from tqdm.notebook import tqdm as tqdm
import pandas as pd

input_dir = "/raid/elhamod/Fish/originals/FMNH"
output_dir = "/raid/elhamod/Fish/cropped2/FMNH"
show_images=False



images_dir = "images"
extends_file = "extends.csv"
skipped_file = "skipped.csv"
index_col = 'Name'
x1_col = 'x1'
x2_col = 'x2'
y1_col = 'y1'
y2_col = 'y2'



df = None

full_output_dir = os.path.join(output_dir, images_dir)
if not os.path.exists(full_output_dir):
    os.makedirs(full_output_dir)
    
csv_full_path = os.path.join(input_dir,extends_file)
extends = pd.read_csv(csv_full_path, delimiter=' ', index_col=index_col)
extends.index = extends.index.map(str.strip)

for fileName in tqdm(os.listdir(os.path.join(input_dir, images_dir))):
    fullName = os.path.join(input_dir, images_dir, fileName)
    try:
        # read
        image = cv2.imread(fullName)
        original = image.copy()

        coord_row = extends.loc[fileName]

        # find region of interest
        x1 = int(coord_row[x1_col])
        y1 = int(coord_row[y1_col])
        x2 = int(coord_row[x2_col])
        y2 = int(coord_row[y2_col])
        w = int(x2 - x1)
        h = int(y2 - y1)
        ROI = original[y1:y1+h, x1:x1+w]

        #Show
        if show_images:
            plt.imshow(ROI)
            plt.show()

        # save the image
        cv2.imwrite(os.path.join(output_dir, images_dir, fileName), ROI)
    except:
        print("image skipped '" + fileName + "'")
        skipped = {
            'fileName': fileName 
        }
        if df is None:
            df = pd.DataFrame()
        df = df.append(pd.DataFrame(skipped, index=[0]), ignore_index = True)
        
if df is not None:
    df.to_csv(os.path.join(input_dir, skipped_file), sep=' ')
```
