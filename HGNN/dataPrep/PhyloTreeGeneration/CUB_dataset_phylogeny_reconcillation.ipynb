```python
import os
import pandas as pd
from Bio import Phylo
from ete3 import Tree

os.environ['QT_QPA_PLATFORM']='offscreen'
```

```python
phlyo_1 = Tree('/raid/elhamod/CUB/official/1_tree-consensus-Hacket-AllSpecies.phy', format=1)
phlyo_1.render('%%inline')
```

```python
species = [node.name for node in phlyo_1.traverse("postorder") if node.name!='']
species
```

```python
taxonomy = pd.read_excel('/raid/elhamod/CUB/official/CUB Taxon reconciliation.xlsx')
taxonomy['English'] = taxonomy['English'].str.replace('-', '_')
taxonomy['English'] = taxonomy['English'].str.replace(' ', '_')

print(len(taxonomy[taxonomy['TipLabel'].isin(species)]))
taxonomy.head(5)
```

```python
# open classes txt file and get the classes for which phylogeny isn't present
with open('/raid/elhamod/CUB/official/CUB_200_2011/classes.txt') as f:
    classes = f.readlines()

classes = [x.split(' ')[1] for x in classes]
classes = [x.split('.')[1] for x in classes]
classes = [x.split('\n')[0] for x in classes]
classes.sort()
classes
```

```python
classes_mapping_present = taxonomy[taxonomy['English'].isin(classes)]
# len(classes_mapping_present)
classes_mapping_present
```

```python
classes_mapping_not_present = taxonomy[~taxonomy['English'].isin(classes)]
len(classes_mapping_not_present)
```

```python
english_to_common_name = pd.read_excel('/raid/elhamod/CUB/official/CUB Taxon reconciliation.xlsx', 'CUB Taxon reconciliation')
english_to_common_name['CUB Common Name (English)'] = english_to_common_name['CUB Common Name (English)'].str.replace(' ', '_')
english_to_common_name['CUB Common Name (English)'] = english_to_common_name['CUB Common Name (English)'].str.replace('_', '_')
english_to_common_name['Birdtree Common Name  (English)'] = english_to_common_name['Birdtree Common Name  (English)'].str.replace(' ', '_')
english_to_common_name['Birdtree Common Name  (English)'] = english_to_common_name['Birdtree Common Name  (English)'].str.replace('-', '_')
english_to_common_name
```

```python
merged = classes_mapping_not_present.merge(english_to_common_name, left_on='English', right_on='Birdtree Common Name  (English)', how='left')
merged.head(2)
```

```python
merged['English'] = merged['CUB Common Name (English)']
merged.drop(columns=['CUB Common Name (English)', 'Birdtree Common Name  (English)', 'Notes', 'Unnamed: 0'], inplace=True)
merged
```

```python
final_reconciliation = classes_mapping_present.append(merged, ignore_index=True)
final_reconciliation.drop(columns='Unnamed: 0', inplace=True)
final_reconciliation.info()
```
